```python
import pandas as pd
import numpy as np
```

```python
class FormulaMagica:
    def __init__(self, patr_liq_ini = 0, patr_liq_fim = 5000000000000):
        self.df = pd.read_clipboard()
        self.df_original = self.df
        self.patr_liq_ini = patr_liq_ini
        self.patr_liq_fim = patr_liq_fim

        self._formata_dados()
        self._filtro()
        self._ordem()
        
    def df_formatado(self):
        print('Total de registros: ' + str(len(self.df)))
        return self.df[['Papel', 'Cotação', 'P/L', 'EV/EBIT', 'ROIC', 'EV/EBIT Ordem', 'ROIC Ordem', 'Ordem Formula']]
        
    def _filtro(self):
        self.df = self.df[(self.df['P/L'] > 1) & (self.df['P/L'] < 1500)]
        self.df = self.df[(self.df['ROIC'] > 0)]
        self.df = self.df[(self.df['EV/EBIT'] > 0)]
        #self.df = self.df[(self.df['Patrim. Líq.'] >= self.patr_liq_ini) & (self.df['Patrim. Líq.'] <= self.patr_liq_fim)]
        
    def _ordem(self):
        # ROIC quanto maior melhor
        # eu ordeno por ordem de ROIC crescente e o primeiro recebe o valor 1 (pior) e o último recebe o maior valor (melhor)
        self.df.sort_values(by='ROIC', inplace=True)
        self.df.insert(3, 'ROIC Ordem', range(1, + len(self.df) + 1))
        
        # quanto maior o EV/EBIT pior
        self.df.sort_values(by='EV/EBIT', inplace=True, ascending=False)
        self.df.insert(3, 'EV/EBIT Ordem', range(1, + len(self.df) + 1))
        
        self.df['Ordem Formula'] = self.df['ROIC Ordem'] + self.df['EV/EBIT Ordem']
        self.df.sort_values(by='Ordem Formula', inplace=True, ascending=False)
        
    def _formata_dados(self):
        self.df['Cotação'] = self.df['Cotação'].str.replace('.', '').str.replace(',', '.').astype('float64')
        self.df['P/L'] = self.df['P/L'].str.replace('.', '').str.replace(',', '.').astype('float64')
        self.df['EV/EBIT'] = self.df['EV/EBIT'].str.replace('.', '').str.replace(',', '.').str.replace('%', '').astype('float64')
        self.df['ROIC'] = self.df['ROIC'].str.replace('.', '').str.replace(',', '.').str.replace('%', '').astype('float64')
        #self.df['Patrim. Líq'] = self.df['Patrim. Líq'].str.replace('.', '').str.replace(',', '.').astype('float64') / 1000
```

```python
# lê os dados do clipboard.
# precisa selecionar a tabela do site: http://www.fundamentus.com.br/resultado.php
df = pd.read_clipboard()
dfOriginal = df
```

```python
# converte os dados para números
df['Cotação'] = df['Cotação'].str.replace(',', '.').astype('float64')
df['P/L'] = df['P/L'].str.replace('.', '').str.replace(',', '.').astype('float64')
df['P/VP'] = df['P/VP'].str.replace('.', '').str.replace(',', '.').astype('float64')
df['PSR'] = df['PSR'].str.replace('.', '').str.replace(',', '.').astype('float64')
df['Div.Yield'] = df['Div.Yield'].str.replace(',', '.').str.replace('%', '').astype('float64') / 100
df['P/Ativo'] = df['P/Ativo'].str.replace('.', '').str.replace(',', '.').astype('float64')
df['P/Cap.Giro'] = df['P/Cap.Giro'].str.replace('.', '').str.replace(',', '.').astype('float64')
df['P/EBIT'] = df['P/EBIT'].str.replace('.', '').str.replace(',', '.').astype('float64')
df['P/Ativ Circ.Liq'] = df['P/Ativ Circ.Liq'].str.replace('.', '').str.replace(',', '.').astype('float64')
df['EV/EBIT'] = df['EV/EBIT'].str.replace('.', '').str.replace(',', '.').astype('float64')
df['Mrg Ebit'] = df['Mrg Ebit'].str.replace('.', '').str.replace(',', '.').str.replace('%', '').astype('float64') / 100
df['Mrg. Líq.'] = df['Mrg. Líq.'].str.replace('.', '').str.replace(',', '.').str.replace('%', '').astype('float64') / 100
df['Liq. Corr.'] = df['Liq. Corr.'].str.replace('.', '').str.replace(',', '.').astype('float64')
df['ROIC'] = df['ROIC'].str.replace('.', '').str.replace(',', '.').str.replace('%', '').astype('float64') / 100
df['ROE'] = df['ROE'].str.replace('.', '').str.replace(',', '.').str.replace('%', '').astype('float64') / 100
df['Liq.2meses'] = df['Liq.2meses'].str.replace('.', '').str.replace(',', '.').astype('float64')
df['Patrim. Líq'] = df['Patrim. Líq'].str.replace('.', '').str.replace(',', '.').astype('float64') / 1000
df['Dív.Brut/ Patrim.'] = df['Dív.Brut/ Patrim.'].str.replace('.', '').str.replace(',', '.').str.replace('%', '').astype('float64') / 100
df['Cresc. Rec.5a'] = df['Cresc. Rec.5a'].str.replace('.', '').str.replace(',', '.').str.replace('%', '').astype('float64') / 100

# filtros
df = df[(df['P/L'] > 1) & (df['P/L'] < 15)]
df = df[(df['ROE'] > 0)]
df = df[(df['Cresc. Rec.5a'] > 0)]
df = df[(df['Liq. Corr.'] > 1)]

# criar valor de acordo com a ordem dos dados
df.sort_values(by='P/L', inplace=True)
df.insert(3, 'P/L Ordem', range(1, + len(df) + 1))
df.sort_values(by='ROE', inplace=True, ascending=False)
df.insert(17, 'ROE Ordem', range(1, + len(df) + 1))

# criar nova coluna com a posição final
df['Ordem Formula'] = df['P/L Ordem'] + df['ROE Ordem']
df.sort_values(by='Ordem Formula', inplace=True)
```
