```python
import pandas as pd
import numpy as np
```

```python
# lê os dados do clipboard.
# precisa selecionar a tabela do site: http://www.fundamentus.com.br/resultado.php
df = pd.read_clipboard()
dfOriginal = df
```

```python
# converte os dados para números
df['Cotação'] = df['Cotação'].str.replace(',', '.').astype('float64')
df['P/L'] = df['P/L'].str.replace(',', '.').astype('float64')
df['P/VP'] = df['P/VP'].str.replace(',', '.').astype('float64')
df['PSR'] = df['PSR'].str.replace(',', '.').astype('float64')
df['Div.Yield'] = df['Div.Yield'].str.replace(',', '.').str.replace('%', '').astype('float64') / 100
df['P/Ativo'] = df['P/Ativo'].str.replace(',', '.').astype('float64')
df['P/Cap.Giro'] = df['P/Cap.Giro'].str.replace(',', '.').astype('float64')
df['P/EBIT'] = df['P/EBIT'].str.replace(',', '.').astype('float64')
df['P/Ativ Circ.Liq'] = df['P/Ativ Circ.Liq'].str.replace(',', '.').astype('float64')
df['EV/EBIT'] = df['EV/EBIT'].str.replace(',', '.').astype('float64')
df['Mrg Ebit'] = df['Mrg Ebit'].str.replace('.', '').str.replace(',', '.').str.replace('%', '').astype('float64') / 100
df['Mrg. Líq.'] = df['Mrg. Líq.'].str.replace(',', '.').str.replace('%', '').astype('float64') / 100
df['Liq. Corr.'] = df['Liq. Corr.'].str.replace(',', '.').astype('float64')
df['ROIC'] = df['ROIC'].str.replace(',', '.').str.replace('%', '').astype('float64') / 100
df['ROE'] = df['ROE'].str.replace(',', '.').str.replace('%', '').astype('float64') / 100
df['Liq.2meses'] = df['Liq.2meses'].str.replace('.', '').str.replace(',', '.').astype('float64')
df['Patrim. Líq'] = df['Patrim. Líq'].str.replace('.', '').str.replace(',', '.').astype('float64') / 1000
df['Dív.Brut/ Patrim.'] = df['Dív.Brut/ Patrim.'].str.replace(',', '.').str.replace('%', '').astype('float64') / 100
df['Cresc. Rec.5a'] = df['Cresc. Rec.5a'].str.replace(',', '.').str.replace('%', '').astype('float64') / 100

# criar valor de acordo com a ordem dos dados
df.sort_values(by='P/L', inplace=True)
df.insert(3, 'P/L Ordem', range(1, + len(df) + 1))
df.sort_values(by='ROE', inplace=True, ascending=False)
df.insert(17, 'ROE Ordem', range(1, + len(df) + 1))

# criar nova coluna com a posição final
df['Ordem Formula'] = df['P/L Ordem'] + df['ROE Ordem']
df.sort_values(by='Ordem Formula', inplace=True)
```
