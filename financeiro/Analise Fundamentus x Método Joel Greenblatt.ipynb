```python
import pandas as pd
import numpy as np
```

```python
# lê os dados do clipboard.
# precisa selecionar a tabela do site: http://www.fundamentus.com.br/resultado.php
ds = pd.read_clipboard()
dsOriginal = ds
```

```python
# converte os dados para números
ds['Cotação'] = ds['Cotação'].str.replace(',', '.').astype('float64')
ds['P/L'] = ds['P/L'].str.replace(',', '.').astype('float64')
ds['P/VP'] = ds['P/VP'].str.replace(',', '.').astype('float64')
ds['PSR'] = ds['PSR'].str.replace(',', '.').astype('float64')
ds['Div.Yield'] = ds['Div.Yield'].str.replace(',', '.').str.replace('%', '').astype('float64') / 100
ds['P/Ativo'] = ds['P/Ativo'].str.replace(',', '.').astype('float64')
ds['P/Cap.Giro'] = ds['P/Cap.Giro'].str.replace(',', '.').astype('float64')
ds['P/EBIT'] = ds['P/EBIT'].str.replace(',', '.').astype('float64')
ds['P/Ativ Circ.Liq'] = ds['P/Ativ Circ.Liq'].str.replace(',', '.').astype('float64')
ds['EV/EBIT'] = ds['EV/EBIT'].str.replace(',', '.').astype('float64')
ds['Mrg Ebit'] = ds['Mrg Ebit'].str.replace('.', '').str.replace(',', '.').str.replace('%', '').astype('float64') / 100
ds['Mrg. Líq.'] = ds['Mrg. Líq.'].str.replace(',', '.').str.replace('%', '').astype('float64') / 100
ds['Liq. Corr.'] = ds['Liq. Corr.'].str.replace(',', '.').astype('float64')
ds['ROIC'] = ds['ROIC'].str.replace(',', '.').str.replace('%', '').astype('float64') / 100
ds['ROE'] = ds['ROE'].str.replace(',', '.').str.replace('%', '').astype('float64') / 100
ds['Liq.2meses'] = ds['Liq.2meses'].str.replace('.', '').str.replace(',', '.').astype('float64')
ds['Patrim. Líq'] = ds['Patrim. Líq'].str.replace('.', '').str.replace(',', '.').astype('float64') / 1000
ds['Dív.Brut/ Patrim.'] = ds['Dív.Brut/ Patrim.'].str.replace(',', '.').str.replace('%', '').astype('float64') / 100
ds['Cresc. Rec.5a'] = ds['Cresc. Rec.5a'].str.replace(',', '.').str.replace('%', '').astype('float64') / 100
```
